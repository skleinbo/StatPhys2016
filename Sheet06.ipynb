# Exercise 16 - DNA-Denaturierung

```julia

y(T, g, 系, N) = g*exp(-系/T)
n(T,g,系,N) = begin
    yy = y(T,g,系,N)
    return N*yy^N/(yy^N-1) - yy/(yy-1)
end
```

```julia
系 = 10*298 # Zehnfache Raumtemperatur

plt = plot(ylabel=L"\langle n \rangle /(N-1)", xlabel="T")
for N in [5, 10, 20, 100]
    plot!(plt, linspace(0,系/5,200), T->n(T, 10^4, 系, N)/(N-1), lab="N=$N")
end
display(plt)
```
